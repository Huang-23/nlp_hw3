```python
import pandas as pd
data = pd.read_csv("movies.csv",encoding = 'utf_8_sig')
data.head()
```

```python
row,col = data.shape
```

```python
data['新分類'] = "create"
for i in range(0,row):
    movie_type = data['分類'][i].replace("[","").replace("]","").replace("\'","").split(',')[0]
    data.loc[i,'新分類'] = movie_type
data
```

```python
data.to_csv( 'movies.csv', encoding='utf_8_sig' , index = False)
```

```python
from sklearn.model_selection import train_test_split

train_data , test_data = train_test_split(data,test_size=0.2)
```

```python
import matplotlib.pyplot as plt

plt.figure( figsize = (20,12))
plt.subplot(1,2,1)
train_data["新分類"].value_counts().sort_index().plot( kind = "barh", title = "訓練集分類")

plt.subplot(1,2,2)
test_data["新分類"].value_counts().sort_index().plot( kind = "barh", title = "測試集分類")
```

```python
import jieba
import re

def content_cut(text):
    specialChars = "！#$%^&*（）《  》【]「"
    text = text.replace("★" , "").replace("\r\n" , "").replace("！","").replace("《","").replace("》","").replace("，","").replace("/","").replace("（","").replace("）","")
    return " ".join(list(jieba.cut(text)))

train_data['分詞劇情'] = train_data['劇情介紹'].map(content_cut)
test_data['分詞劇情'] = test_data['劇情介紹'].map(content_cut)
```

```python
train_data
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer = TfidfVectorizer(sublinear_tf=False, stop_words=None, token_pattern="(?u)\\b\\w+\\b", smooth_idf=True, norm='l2')
#vectorizer = CountVectorizer(stop_words=None, token_pattern="(?u)\\b\\w+\\b")
X_train = vectorizer.fit_transform(train_data["分詞劇情"])
X_test = vectorizer.transform(test_data["分詞劇情"])
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=22)
knn.fit(X_train, train_data["新分類"])


Y_test = knn.predict(X_test)

assess_model_socre=knn.score(X_test,test_data["新分類"])
print('Test set score:{:2f}'.format(assess_model_socre))
```
